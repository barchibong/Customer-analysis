```python
#creating the price_label column
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#checking the value counts
df['price_range'].value_counts(dropna = False)
```

```python
#checking the maximum value within the price column
df['prices']. max ()
```

```python
#Using the loc function on a subset of my data frame
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#creating the busiest day of the week column
ords_prods_merged['order_dow'].value_counts(dropna = False)
```

```python
#From the output, 0 which refers to Saturday has the highest frequency which makes it the busiest day and 4 which refers to Wednesday is slowest day of the week when it comes to instacart orders.
```

```python
#Creating a new column "busiest day" that will contain busiest day, least day and regularly busy.
result = []

for value in ords_prods_merged["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
ords_prods_merged['busiest_day'] = result
```

```python
ords_prods_merged['busiest_day']. value_counts (dropna = False)
```

```python
#Creating a new column "busiest days" that will contain the two busiest days of the week.
result = []

for value in ords_prods_merged["order_dow"]:
  if value == 0:
    result.append("First busiest")
  elif value == 1:
    result.append("Second busiest")
  else:
    result.append("Regularly busy")
  
```
