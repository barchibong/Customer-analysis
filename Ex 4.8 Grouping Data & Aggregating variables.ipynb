```python
#Importing Libraries
import pandas as pd
import numpy as np
import os
```

```python
#Combine the pickle files back into a single DataFrame:
# Initialize an empty list to hold the individual DataFrames from the pickle files
dfs = []
```

```python
num_chunks = 5
```

```python
# Load each pickle file and append its DataFrame to the list
for i in range(num_chunks):
    chunk_df = pd.read_pickle(f'chunk_{i}.pickle')
    dfs.append(chunk_df)
```

```python
# Concatenate the DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
ords_prods_merge = combined_df
```

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Creating a loyalty flag
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Calculate the number of rows in each chunk
chunk_size = len(ords_prods_merge) // num_chunks
```

```python
# Split the DataFrame into chunks
chunks = [ords_prods_merge[i:i+chunk_size] for i in range(0, len(ords_prods_merge), chunk_size)]
```
