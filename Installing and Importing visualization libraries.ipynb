```python
#Importing prepared data 
orders_products_merged = pd.read_pickle(r'C:\Users\B.Archibong\Downloads\Instacart Basket Analysis\02 Data\Prepared Data\orders_products_merged.pkl')
```

```python
#COMBINING CUSTOMER DATA WITH PREPARED INSTACART DATA
frames = [df_customerdata, orders_products_merged]
```

```python
df_concat = pd.concat (frames)
```

```python
df_concat
```

```python
num_chunks = 5
```

```python
# Calculate the number of rows in each chunk
chunk_size = len(df_concat) // num_chunks
```

```python
# Split the DataFrame into chunks
chunks = [df_concat[i:i+chunk_size] for i in range(0, len(df_concat), chunk_size)]
```
